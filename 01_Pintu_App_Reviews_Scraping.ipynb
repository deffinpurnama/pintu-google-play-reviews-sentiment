```python
# Mengimpor pustaka google_play_scraper untuk mengakses ulasan dan informasi aplikasi dari Google Play Store.
from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd
```

# Scrapping Dataset


```python
# Mengambil semua ulasan dari aplikasi Pintu
ulasan_pintu = reviews_all(
    'com.valar.pintu',        # ID aplikasi Pintu
    lang='id',                # Bahasa ulasan: Indonesia
    country='id',             # Negara: Indonesia
    sort=Sort.MOST_RELEVANT,  # Urutan ulasan: Paling relevan
    sleep_milliseconds=0,     # Waktu tunggu antar permintaan
)
```

```python
# Ubah hasil ulasan menjadi DataFrame pandas
df = pd.DataFrame(ulasan_pintu)

# Simpan ke file CSV
df.to_csv('ulasan_pintu.csv', index=False, encoding='utf-8-sig')

print("Ulasan berhasil disimpan ke file 'ulasan_pintu.csv'")
```
